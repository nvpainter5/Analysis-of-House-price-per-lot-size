```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (8.0, 8.0)
plt.style.use('ggplot')
df = pd.read_csv('Housing.csv')
df.plot(kind='scatter',x='lotsize', y='price', c = 'blue')
print('The relationship shows that its non-linear. Its definiely in positive direction. And I feel it is week strength.')
print('There are 15 outliers (calculated in the excel), these outliers also influence a non linear relation between price and lot size.')
```

```python
from IPython.display import Image
Image(filename='outliers.png') 
```

```python
from scipy.stats import linregress

x = df['lotsize']
y = df['price']
slope, intercept, r_value, p_value, std_err = linregress(x,y)
plt.plot(x, y, 'o', label='original data', color='orange')
plt.plot(x, intercept + slope*x, 'r', label='fitted line')
plt.title(f"Least - Squares Regression Line {round(intercept, 2)}+{round(slope,2)}*x")
plt.xlabel('LotSize')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
predicted_value_1 = float(intercept) + float(slope)*5000
predicted_value_2 = float(intercept) + float(slope)*7500
predicted_value_3 = float(intercept) + float(slope)*10000
print(predicted_value_1)
print(predicted_value_2)
print(predicted_value_3)
```

```python
import numpy as np
import seaborn as sns
import pandas as pd

sns.set(style="whitegrid")
df = pd.read_csv('Housing.csv')

rs = np.random.RandomState(7)
x = rs.normal(df.price)
y = 2 + 1.5 * x + rs.normal(df.lotsize)
sns.residplot(x, y, lowess=True, color='green')
print('The residuals are negative and are non linear/underfitted.')
```

```python
actual_value = df.loc[df['lotsize'] == 5000]
residual_1 = actual_value['price'] - predicted_value_1
print('Below is Residual for 5000:')
print(residual_1)

print('Below is Residual for 7500:')
actual_value = df.loc[df['lotsize'] == 7500]
residual_2 = actual_value['price'] - predicted_value_2
print(residual_2)

print('Below is Residual for 10000:')
actual_value = df.loc[df['lotsize'] == 10000]
residual_3 = actual_value['price'] - predicted_value_3
print(residual_3)
print('There are no residuals for 10000 lot size because there is no 10000 actual value in the dataset.')
```
